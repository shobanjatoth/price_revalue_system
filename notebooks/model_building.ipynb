```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error,r2_score 
from xgboost import XGBRegressor
from sklearn.linear_model import SGDRegressor
```

```python
df=pd.read_csv("D:\\price_revalue_stystem\\data\\04-predictionsfeature_selected.csv")
```

```python
cols_to_keep = [
    "Type_HatchBack",
    "age_of_car",
    "model_count",
    "Year",
    "brand_count",
    "Type_Sedan",
    "km_per_year",
    "Distance",
    "Drive_Automatic",
    "Drive_Manual",
    "Owner",
    "Fuel_Diesel",
    "Fuel_Petrol",
    "Type_SUV",
    "Price_log"
]

df = df[cols_to_keep]
```

```python
x=df.drop('Price_log', axis=1)
y=df['Price_log']
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

# Sgd_regressor

```python
sgd=SGDRegressor(max_iter=1000, tol=1e-3)
sgd.fit(x_train, y_train)
y_pred_sgd=sgd.predict(x_test)
mse_sgd=mean_squared_error(y_test, y_pred_sgd)
rmse=np.sqrt(mse_sgd)
r2_sgd=r2_score(y_test, y_pred_sgd)
print(f"SGD Regressor RMSE: {rmse}")
print(f"SGD Regressor MSE: {mse_sgd}")
print(f"SGD Regressor R^2 Score: {r2_sgd}")
t=sgd.predict(x_train)
mse_sgd_train=mean_squared_error(y_train, t)
print(f"SGD Regressor Training MSE: {mse_sgd_train}")
print(f"SGD Regressor Training RMSE: {np.sqrt(mse_sgd_train)}")
print(f"SGD Regressor Training R^2 Score: {r2_score(y_train, t)}")
```

# Randomforest

```python
rf=RandomForestRegressor(n_estimators=100, random_state=42)
rf.fit(x_train, y_train)
y_pred_rf=rf.predict(x_test)
mse_rf=mean_squared_error(y_test, y_pred_rf)
rmse_rf=np.sqrt(mse_rf)
r2_rf=r2_score(y_test, y_pred_rf)
print(f"Random Forest Regressor RMSE: {rmse_rf}")
print(f"Random Forest Regressor MSE: {mse_rf}")
print(f"Random Forest Regressor R^2 Score: {r2_rf}")
t_rf=rf.predict(x_train)
mse_rf_train=mean_squared_error(y_train, t_rf)
print(f"Random Forest Regressor Training MSE: {mse_rf_train}")
print(f"Random Forest Regressor Training RMSE: {np.sqrt(mse_rf_train)}")
print(f"Random Forest Regressor Training R^2 Score: {r2_score(y_train, t_rf)}")
```

# xgboost

```python
xgb=XGBRegressor(n_estimators=100, random_state=42)
xgb.fit(x_train, y_train)
y_pred_xgb=xgb.predict(x_test)
mse_xgb=mean_squared_error(y_test, y_pred_xgb)
rmse_xgb=np.sqrt(mse_xgb)
r2_xgb=r2_score(y_test, y_pred_xgb)
print(f"XGBoost Regressor RMSE: {rmse_xgb}")
print(f"XGBoost Regressor MSE: {mse_xgb}")
print(f"XGBoost Regressor R^2 Score: {r2_xgb}")
t_xgb=xgb.predict(x_train)
mse_xgb_train=mean_squared_error(y_train, t_xgb)
print(f"XGBoost Regressor Training MSE: {mse_xgb_train}")
print(f"XGBoost Regressor Training RMSE: {np.sqrt(mse_xgb_train)}")
print(f"XGBoost Regressor Training R^2 Score: {r2_score(y_train, t_xgb)}")
# rf_importance=pd.Series(
#     rf.feature_importances_, index=x_train.columns)
# rf_importance.sort_values(ascending=False, inplace=True)
# rf_importance.head(15)
```

```python


def gap_analysis(model, X_train, y_train, X_test, y_test):
    
    # Predictions
    y_train_pred = model.predict(X_train)
    y_test_pred = model.predict(X_test)
    
    # Metrics
    train_rmse = np.sqrt(mean_squared_error(y_train, y_train_pred))
    test_rmse = np.sqrt(mean_squared_error(y_test, y_test_pred))
    
    train_r2 = r2_score(y_train, y_train_pred)
    test_r2 = r2_score(y_test, y_test_pred)
    
    # Gaps
    rmse_gap = test_rmse - train_rmse
    r2_gap = train_r2 - test_r2
    
    rmse_gap_percent = (rmse_gap / train_rmse) * 100
    
    results = pd.DataFrame({
        "Metric": ["RMSE", "R2"],
        "Train": [train_rmse, train_r2],
        "Test": [test_rmse, test_r2],
        "Gap": [rmse_gap, r2_gap],
        "Gap_%": [rmse_gap_percent, None]
    })
    
    return results
gap_results = gap_analysis(xgb, x_train, y_train, x_test, y_test)
print(gap_results)
```

# cross validation

```python
from sklearn.model_selection import cross_val_score
cv_scores = cross_val_score(xgb, x, y, cv=5, scoring='neg_mean_squared_error')
cv_rmse_scores = np.sqrt(-cv_scores)
print(f"Cross-Validation RMSE Scores: {cv_rmse_scores}")
print(f"Average CV RMSE: {cv_rmse_scores.mean()}")
print(f"Average CV R^2 Score: {cross_val_score(xgb, x, y, cv=5, scoring='r2').mean()}")
print(f"std: {cv_rmse_scores.std()}")
```

```python
from sklearn.model_selection import StratifiedKFold
import numpy as np

y_bins = pd.qcut(y, q=5, labels=False)

skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

scores = []

for train_idx, val_idx in skf.split(x, y_bins):
    X_tr, X_val = x.iloc[train_idx], x.iloc[val_idx]
    y_tr, y_val = y.iloc[train_idx], y.iloc[val_idx]

    xgb.fit(X_tr, y_tr)
    pred = xgb.predict(X_val)
    rmse = np.sqrt(mean_squared_error(y_val, pred))
    scores.append(rmse)

print("Stratified CV RMSE:", scores)
print("Average:", np.mean(scores))
print("Std:", np.std(scores))
```
